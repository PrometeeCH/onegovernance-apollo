```python
import numpy as np
import pandas as pd

file_path = "../../../data/data_gen/cleaned_project_data.csv"
df = pd.read_csv(file_path)

# Supposons que votre DataFrame s'appelle df et que la colonne concernée est 'DateColumn'
start_date = pd.to_datetime("2024-12-02")
end_date = pd.to_datetime("2026-07-08")

# Calculer le nombre total de jours entre les deux dates
total_days = (end_date - start_date).days

# Identifier les indices des NaN dans la colonne
nan_indices = df["10. End date_analyse"].isna()

# Compter le nombre de NaN
num_nans = nan_indices.sum()

# Générer des nombres aléatoires de jours à ajouter à la date de début
random_days = np.random.randint(0, total_days + 1, num_nans)

# Créer les dates aléatoires sans inclure les heures
random_dates = [start_date + pd.Timedelta(days=int(day)) for day in random_days]

# Convertir les dates au format désiré sans heure (YYYY-MM-DD)
random_dates_formatted = [date.strftime("%Y-%m-%d") for date in random_dates]

# Remplacer les NaN dans la colonne par les dates aléatoires formatées
df.loc[nan_indices, "10. End date_analyse"] = random_dates_formatted
display(df.head(10))
```
